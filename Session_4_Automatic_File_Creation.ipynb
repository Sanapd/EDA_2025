```python
import os 
folder_name='bar_charts'  # change1
try:
    os.makedirs(folder_name)
except Exception as e:
    print(e)
cwd=os.getcwd()
for i in cat[1:]:
    keys=visa_df[i].value_counts().keys()
    values=visa_df[i].value_counts().values
    plt.bar(keys,values) # chang2
    file_name=f"{i}.jpg"
    path=os.path.join(cwd,folder_name,file_name)
    plt.savefig(path)
```

**Pie-Charts**

```python
import os 
folder_name='Pie_charts'
try:
    os.makedirs(folder_name)
except Exception as e:
    print(e)
cwd=os.getcwd()
for i in cat[1:]:
    keys=visa_df[i].value_counts().keys()
    values=visa_df[i].value_counts().values
    plt.pie(values,
            labels=keys,
            autopct='%0.2f%%')
    file_name=f'{i}.jpg'
    path=os.path.join(cwd,folder_name,file_name)
    plt.savefig(path)
```
