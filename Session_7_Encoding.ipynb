**LableEncoder vs OneHotEncoder**

- Generally we dont want a relation between input columns

- We need both input columns are **not correlated** to each other

- We need both input columns are **Independent(X,Y)** to each other

- We need both input columns are **Perpendicular** to each other

- We need both input columns are **Orthogonal** to each other

- We need both input columns are **90 degrees phase shift** to each other

- OneHotEncoder will achieve above

- But OneHotEncoder creates new columns
  - columns are increase means dimension are increase

  - More dimension data need more processing time

  - More dimension data need more infra

  - This is called Curse Of Dimension
